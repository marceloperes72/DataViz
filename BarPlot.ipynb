```python
def bar_plot(x, y, title_label, y_label,destaque,destaque_texto ,cor_barras,cor_destaque):
    """"Plota um gráfico de barras com valor sobre a ba
    Args:
          x: eixo x.
          y: eixo y.
          title_label(str): título do gráfico.
          y_label(str): Nome do label do eixo Y.
          destaque(int): qual barra você quer destacar
          destaque_texto(str): texto para anotar sobre o gráfico ,pode ser " " caso não seja necessário
          cor_barras:cor das barras
          cor_destaque:cor do destaque
    """
    # Determina o estilo
    plt.style.use('bmh')
    plt.rcParams['axes.grid'] = False

    # Determina a cor de fundo do gráfico
    plt.rcParams['axes.facecolor'] = 'white'    
    # Determina o tamanho do gráfico.
    plt.figure(figsize=(8,5), dpi=100)
    #Plota o título
    plt.title(title_label, fontsize=18)
    #Plota o label do eixo y
    plt.ylabel(y_label,fontsize=16, horizontalalignment='left', y=0)
    #ordena as barras de forma decrescente
    y,x = zip(*sorted(zip(y,x),reverse=True))
    #plota os valores de cada barra
    barplot = plt.bar(x, y,color=cor_barras)
    #plota a barra destacada
    barplot[destaque].set_color(cor_destaque)
    plt.text(barplot[destaque].get_x() + barplot[destaque].get_width()/100,barplot[destaque].get_height()+20 ,destaque_texto, va='bottom',size =14, color = cor_destaque) 
    for bar in barplot:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/4, yval - 30, int(yval), va='top',size =12, color = 'white') 
        #va: seleciona onde plota o texto  
    #seta a base do gráfico com 0
    plt.ylim(ymin=0)  
    #elimina a caixa que delimita o gráfico
    plt.box(on=None)
    #Mostra o gráfico
    plt.show()
```

```python
def bar_plot_h(x, y, title_label, y_label,destaque,destaque_texto ,cor_barras,cor_destaque):
    """"Plota um gráfico de barras com valor sobre a ba
    Args:
          x: eixo x.
          y: eixo y.
          title_label(str): título do gráfico.
          y_label(str): Nome do label do eixo Y.
          destaque(int): qual barra você quer destacar
          destaque_texto(str): texto para anotar sobre o gráfico ,pode ser " " caso não seja necessário
          cor_barras:cor das barras
          cor_destaque:cor do destaque
    """
    # Determina o estilo
    plt.style.use('bmh')
    # Determina a cor de fundo do gráfico
    plt.rcParams['axes.facecolor'] = 'white'    
    plt.rcParams['axes.grid'] = False
    # Determina o tamanho do gráfico.
    plt.figure(figsize=(8,5), dpi=100)
    #Plota o título
    plt.title(title_label, fontsize=18)
    #Plota o label do eixo y
    plt.xlabel(y_label,fontsize=16, x= 0.05)
    #ordena as barras de forma decrescente
    y,x = zip(*sorted(zip(y,x)))
    #plota os valores de cada barra
    barplot = plt.barh(x, y,color=cor_barras)
    #plota a barra destacada
    barplot[destaque].set_color(cor_destaque)
    plt.text(barplot[destaque].get_x() + barplot[destaque].get_width() +20, barplot[destaque].get_y() +0.25 * barplot[destaque].get_height() , destaque_texto, va='bottom', size =14, color = cor_destaque) 
    for bar in barplot:
        yval = bar.get_width()
        plt.text(bar.get_width() - 80, bar.get_y() + 0.5 * bar.get_height(), int(yval), ha='center', va='center',size =12, color = 'white') 
        #va: seleciona onde plota o texto  
    #elimina a caixa que delimita o gráfico
    plt.box(on=None)
    #Mostra o gráfico
    plt.show()
```
